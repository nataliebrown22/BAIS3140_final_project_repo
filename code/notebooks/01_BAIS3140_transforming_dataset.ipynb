#### Transform Data to Keep Rows where Arsenal is Team and Determine whether They are Home or Away

```python
# Create a DataFrame for home games with 'Opponent' column
home_df = soccer_df.rename(columns={'Home Team': 'Team', 'Away Team': 'Opponent'})
home_df['Home/Away'] = 'Home'

# Create a DataFrame for away games with 'Opponent' column
away_df = soccer_df.rename(columns={'Away Team': 'Team', 'Home Team': 'Opponent'})
away_df['Home/Away'] = 'Away'

# Combine the dataframes
melted_df = pd.concat([home_df, away_df], ignore_index=True)

# Review the result
melted_df
```

```python
# rename half time result to ahead, behind, tied
def calculate_half_time_result(row):
    if row['Home/Away'] == 'Home':
        if row['Half Time Home Team Goals'] > row['Half Time Away Team Goals']:
            return 'Ahead'
        elif row['Half Time Home Team Goals'] < row['Half Time Away Team Goals']:
            return 'Behind'
        elif row['Half Time Home Team Goals'] == row['Half Time Away Team Goals']:
            return 'Tied'
    elif row['Home/Away'] == 'Away':
        if row['Half Time Home Team Goals'] < row['Half Time Away Team Goals']:
            return 'Ahead'
        elif row['Half Time Home Team Goals'] > row['Half Time Away Team Goals']:
            return 'Behind'
        elif row['Half Time Home Team Goals'] == row['Half Time Away Team Goals']:
            return 'Tied'

melted_df['Half Time Result']=melted_df.apply(calculate_half_time_result, axis=1)

melted_df
```

```python
# do the same for the match result
def calculate_half_time_result(row):
    if row['Home/Away'] == 'Home':
        if row['Full Time Home Team Goals'] > row['Full Time Away Team Goals']:
            return 'Win'
        elif row['Full Time Home Team Goals'] < row['Full Time Away Team Goals']:
            return 'Loss'
        elif row['Full Time Home Team Goals'] == row['Full Time Away Team Goals']:
            return 'Draw'
    elif row['Home/Away'] == 'Away':
        if row['Full Time Home Team Goals'] < row['Full Time Away Team Goals']:
            return 'Win'
        elif row['Full Time Home Team Goals'] > row['Full Time Away Team Goals']:
            return 'Loss'
        elif row['Full Time Home Team Goals'] == row['Full Time Away Team Goals']:
            return 'Draw'

melted_df['Game Result']=melted_df.apply(calculate_half_time_result, axis=1)

melted_df
```

```python
cols=melted_df.columns
for col in cols:
    print(f'\'{col}\',')
```

```python
# convert columns

melted_df['Total Goals For']=melted_df.apply(
    lambda row: row['Full Time Home Team Goals'] if row['Home/Away'] == 'Home' else row['Full Time Away Team Goals'],axis=1)

melted_df['Total Goals Against']=melted_df.apply(
    lambda row: row['Full Time Away Team Goals'] if row['Home/Away'] == 'Home' else row['Full Time Home Team Goals'],axis=1)

melted_df['Half Time Goals For']=melted_df.apply(
    lambda row: row['Half Time Home Team Goals'] if row['Home/Away'] == 'Home' else row['Half Time Away Team Goals'],axis=1)

melted_df['Half Time Goals Against']=melted_df.apply(
    lambda row: row['Half Time Away Team Goals'] if row['Home/Away'] == 'Home' else row['Half Time Home Team Goals'],axis=1)

melted_df['Shots For']=melted_df.apply(
    lambda row: row['Home Team Shots'] if row['Home/Away'] == 'Home' else row['Away Team Shots'],axis=1)

melted_df['Shots Against']=melted_df.apply(
    lambda row: row['Away Team Shots'] if row['Home/Away'] == 'Home' else row['Home Team Shots'],axis=1)

melted_df['Shots For On Goal']=melted_df.apply(
    lambda row: row['Home Team Shots on Goal'] if row['Home/Away'] == 'Home' else row['Away Team Shots on Goal'],axis=1)

melted_df['Shots Against On Goal']=melted_df.apply(
    lambda row: row['Away Team Shots on Goal'] if row['Home/Away'] == 'Home' else row['Home Team Shots on Goal'],axis=1)

melted_df['Corners For']=melted_df.apply(
    lambda row: row['Home Team Corners'] if row['Home/Away'] == 'Home' else row['Away Team Corners'],axis=1)

melted_df['Corners Against']=melted_df.apply(
    lambda row: row['Away Team Corners'] if row['Home/Away'] == 'Home' else row['Home Team Corners'],axis=1)

melted_df['Yellow Cards']=melted_df.apply(
    lambda row: row['Home Team Yellow Cards'] if row['Home/Away'] == 'Home' else row['Away Team Yellow Cards'],axis=1)

melted_df['Oppenent Yellow Cards']=melted_df.apply(
    lambda row: row['Away Team Yellow Cards'] if row['Home/Away'] == 'Home' else row['Home Team Yellow Cards'],axis=1)

melted_df['Fouls Committed']=melted_df.apply(
    lambda row: row['Home Team Fouls Committed'] if row['Home/Away'] == 'Home' else row['Away Team Fouls Committed'],axis=1)

melted_df['Oppenent Fouls Committed']=melted_df.apply(
    lambda row: row['Away Team Fouls Committed'] if row['Home/Away'] == 'Home' else row['Home Team Fouls Committed'],axis=1)

melted_df['Red Cards']=melted_df.apply(
    lambda row: row['Home Team Red Cards'] if row['Home/Away'] == 'Home' else row['Away Team Red Cards'],axis=1)

melted_df['Oppenent Red Cards']=melted_df.apply(
    lambda row: row['Away Team Red Cards'] if row['Home/Away'] == 'Home' else row['Home Team Red Cards'],axis=1)

cols_to_drop=['Full Time Home Team Goals','Full Time Away Team Goals','Half Time Home Team Goals','Half Time Away Team Goals',
              'Half Time Result','Home Team Shots','Away Team Shots','Home Team Shots on Goal','Away Team Shots on Goal',
              'Home Team Fouls Committed','Away Team Fouls Committed','Home Team Corners','Away Team Corners',
              'Home Team Yellow Cards','Away Team Yellow Cards','Home Team Red Cards','Away Team Red Cards']

melted_df=melted_df.drop(columns=cols_to_drop)

melted_df.head()






```

```python
# Convert 'Match Date' to datetime 
melted_df['Match Date'] = pd.to_datetime(melted_df['Match Date'], format='%d/%m/%y') 
# Filter out rows where the year is 2019 
melted_df = melted_df[melted_df['Match Date'].dt.year != 2019]

melted_df.head()
```
