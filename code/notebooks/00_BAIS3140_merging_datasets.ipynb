```python
import pandas as pd
```

```python
def read_csv(fh,path):
    handle=path+fh
    df=pd.read_csv(handle,encoding='utf-8',sep=',')
    return df

# function for dataframe shape, returns shape
def df_shape(df,name):
    rows=df.shape[0]
    cols=df.shape[1]
    print(f'{name} Data Frame\n{'-'*25}\nRows: {rows}\nColumns: {cols}\n\n')

def df_columns(df,name):
    cols=df.columns
    print(f'{name} Columns\n{'-'*20}\n')
    for col in cols:
        print(col)

def check_columns_match(dataframes):
    if not dataframes:
        return 'Error, no dataframes to compare'  # Handling case of an empty list
    
    first_df_columns = dataframes[0].columns
    for df in dataframes[1:]:
        if not df.columns.equals(first_df_columns):
            return 'Error, columns do not match'
    return 'Success, all columns match'

# function for nulls, returns a dataframe with nulls
def df_nulls(df):
    null_counts=df.isna().sum()
    null_percentages=round((df.isna().sum()/df.shape[0])*100,2)

    # data frame for results
    nulls_df=pd.DataFrame({
        'Nulls':null_counts,
        'Nulls (%)':null_percentages
    })

    return nulls_df

def datatypes(df):
    print(df.dtypes)
```

```python
path = 'C:\\Users\\namb2\\OneDrive\\Documents\\GitHub\\BAIS3140_final_project_repo\\data\\raw_data\\'

season_9_10=read_csv('season-0910.csv',path)
season_10_11=read_csv('season-1011.csv',path)
season_11_12=read_csv('season-1112.csv',path)
season_12_13=read_csv('season-1213.csv',path)
season_13_14=read_csv('season-1314.csv',path)
season_14_15=read_csv('season-1415.csv',path)
season_15_16=read_csv('season-1516.csv',path)
season_16_17=read_csv('season-1617.csv',path)
season_17_18=read_csv('season-1718.csv',path)
season_18_19=read_csv('season-1819.csv',path)
season_19_20=read_csv('season-1920.csv',path)
season_20_21=read_csv('season-2021.csv',path)
season_21_22=read_csv('season-2122.csv',path)
season_22_23=read_csv('season-2223.csv',path)
season_23_24=read_csv('season-2324.csv',path)
```

```python
dataframes=[season_9_10,season_10_11,season_11_12,season_12_13,season_13_14,
            season_14_15,season_15_16,season_16_17,season_17_18,season_18_19,season_19_20,
           season_20_21,season_21_22,season_22_23,season_23_24]
names=["09-10", "10-11", "11-12", "12-13", "13-14", "14-15", "15-16", "16-17", "17-18", "18-19",'19-20','20-21','21-22','22-23','23-24']


for df,name in zip(dataframes,names):
    df_shape(df,name)
```

```python
# view columns for each
for df,name in zip(dataframes,names):
    df_columns(df,name)
    print(f'{'-'*20}\n\n')
```

```python
# check if they match
check_columns_match(dataframes)
```

```python
# continue to vertically merge
seasons_09_24=pd.concat(dataframes,ignore_index=True)

# review
seasons_09_24.head()
```

```python
# check for nulls
seasons_09_24_nulls_df=df_nulls(seasons_09_24)

seasons_09_24_nulls_df
```
